Подготовка:

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
pd.set_option('display.max_rows', None)
from IPython.display import display

sale = pd.read_csv('sale_of_tour_packages.csv', delimiter=';', encoding='utf8')
agent = pd.read_csv('travel_agents.csv', delimiter=';')
travel = pd.read_csv('travels.csv', delimiter=';')

temp = pd.merge(sale, agent)
all_tours = pd.merge(temp, travel)
```

1 задание:

```python
cities = ['Москва', 'Вологда', 'Мурманск', 'Петрозаводск', 'Санкт-Петербург', 
          'Ярославль', 'Тверь', 'Иваново', 'Кострома', 'Архангельск', 'Псков', 
          'Сыктывкар']
task1_temp = all_tours[all_tours['Город'].isin(cities)]
task1 = task1_temp.groupby('Город')['Количество проданных путёвок'].sum().sum()
 
print('Количество человек, совершивших путешествие в региональные центры:')
display(task1)
print('')
```

Задание 2:

```python
gorizont = all_tours[all_tours['Название'] == 'Горизонт']
task2_1 = gorizont.groupby('Город')['Количество проданных путёвок'].sum()
task2_2 = gorizont.groupby('Город')['Стоимость, на 1 чел'].sum()
task2 = task2_1 * task2_2

pie, ax = plt.subplots(figsize=[10,6])
labels = task2.keys()
plt.pie(x=task2, labels=labels, autopct="%.1f%%", pctdistance=0.8, 
        textprops={'fontsize': 8})
plt.title("Туроператор Горизонт. Стоимость путёвок", fontsize=14)

display(task2)
total = task2.sum()
print("Общая стоимость путёвок:", total)

plt.show()
print('')
```

Задание 3:

```python
bar, ax = plt.subplots(figsize=[10,6])
gorizont = all_tours[all_tours['Название'] == 'Мечта']
task3 = gorizont.groupby('Дата')['Количество проданных путёвок'].sum()
labels = task3.keys()
plt.tick_params (axis = 'x', labelsize = 8)
plt.bar(x=labels, height=task3)
plt.xticks(rotation = 'vertical')
plt.title("Туроператор Мечта. Количество проданных путевок в каждый из дней", 
          fontsize=14)
plt.show()
```
